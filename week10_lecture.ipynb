### the absolute value... the easiest function in maths which make things the most confusing!

$ |12| = 12 $ but $ |-12| = 12 $

define the fucntion $f(x)=||x|-1|$ and plot $f(x)$ and $f(f(x))$ and $f(f(f(x)))$

```python
import sympy
def f(x):
    return sympy.Abs((sympy.Abs(x)-1))
    
sympy.plotting.plot(f(x), f(f(x)), f(f(f(x))), (x, -10*sympy.pi, 10*sympy.pi), legend=True)
```

## Calculus

solve the equation $ x^2 - 1 = 0 $

we have $x^2-1 = (x+1)(x-1)$ so either $x+1=0$ or $x-1=0$ ie $x=1$ or $x=-1$

```python
sympy.solvers.solve(x**2-1, x)
sympy.plotting.plot(x**2-1, (x,-2,2))
```

solve the equation $x^3-3x^2+2x-1=0$

order of 3, so gauranteed to have 3 solutions

```python
sympy.solvers.solve(((x**3) - (3*x**2) + (2*x) - 1), x)
```

```python
sympy.plotting.plot(((x**3) - (3*x**2) + (2*x) - 1), (x,-2,4))
# there is one "real" solution, the one that is y-intercept (crosses y axis zero)
```

investigate that there are 11 solutions to the equation $2cos^2(2x)-sin(x/2)=1$ in the interval $[0,3\pi]$

note $\cos^2(x)$ this means $(cos(x))^2$

```python
def f(x):
    return ((2*sympy.cos(2*x)**2) - (sympy.sin(x/2))-1)

sympy.solvers.solve(f(x), x)
```

```python
sympy.plotting.plot( (2*sympy.cos(2*x)**2) - (sympy.sin(x/2))-1, (x, 0, 3*sympy.pi))
```
